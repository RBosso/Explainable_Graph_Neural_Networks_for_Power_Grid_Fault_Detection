```python
%%time

import py_dss_interface
import pandas as pd 
import numpy as np
import os

BusFile = os.getcwd() + "/34Bus/IEEE34_BusXY.csv"
BusFile = BusFile.replace("\\","/")
print(BusFile)
Buses = pd.read_csv(BusFile, header=None)
Buses = Buses[0]
Buses = Buses.to_numpy()
print(Buses)

Experiment = ["New Fault.F1 phases="," r="]
FaultResistances = [0.0001]
FaultLabels = [[["3 bus1="," bus2=",".4.4.4"],"Three Phase to Neutral"], #Three Phase Fault to neutral
[["3 bus1="," bus2=",".0.0.0"],"Three Phase to Ground"], #Three Phase Fault to ground (this should work)
[["1 bus1=",".1 bus2=",".0"],"Single Line to Ground"], #Single Line-to-ground fault
[["1 bus1=",".2 bus2=",".0"],"Single Line to Ground"], #Single Line-to-ground fault
[["1 bus1=",".3 bus2=",".0"],"Single Line to Ground"], #Single Line-to-ground fault
# [["1 bus1=",".3 bus2=",".4"],"Single Line to Neutral"], #Single Line-to-neutral fault
[["1 bus1=",".2 bus2=",".3"],"Single Line to Line"], #Single Line-to-line fault
[["1 bus1=",".1 bus2=",".2"],"Single Line to Line"], #Single Line-to-line fault
[["2 bus1=",".2.3 bus2=",".0.0"],"Two Line to Ground"],#Two Line-to-ground fault
[["2 bus1=",".1.3 bus2=",".0.0"],"Two Line to Ground"],#Two Line-to-ground fault (Not sure if this will work)
[["2 bus1=",".1.2 bus2=",".0.0"],"Two Line to Ground"]]

FaultLabelSimulations = []

dss_file = os.getcwd() + "/34Bus/ieee34Mod1.dss"
dss_file = dss_file.replace("\\","/")
print(dss_file)

FaultLabelSimulations = []

for bus in Buses:
    for fault in FaultLabels:
        for ohm in FaultResistances:
            # print(Experiment[0]+fault[0]+ bus + fault[1] + bus + fault[2]+Experiment[-1]+str(ohm))
            FaultLabelSimulations.append([Experiment[0]+fault[0][0]+ bus + fault[0][1] + bus + fault[0][2]+Experiment[-1]+str(ohm), fault[1]])
print(len(FaultLabelSimulations))
# print(FaultLabelSimulations)

#Initiializes the FaultSimulationData array
FaultSimulationData16 = np.array([['Bus', ' BasekV', ' Node1', ' Magnitude1', ' Angle1', ' pu1', ' Node2',
       ' Magnitude2', ' Angle2', ' pu2', ' Node3', ' Magnitude3', ' Angle3',
       ' pu3', 'Fault_Command',"PV_Irradiance","Fault_Type",'CircuitSolved']])

FaultSimulationData20 = np.array([['Bus', ' BasekV', ' Node1', ' Magnitude1', ' Angle1', ' pu1', ' Node2',
       ' Magnitude2', ' Angle2', ' pu2', ' Node3', ' Magnitude3', ' Angle3',
       ' pu3', ' Node4', ' Magnitude4', ' Angle4', ' pu4', 'Fault_Command',"PV_Irradiance","Fault_Type",'CircuitSolved']])

IrradianceLoop = [0, 0.2, 0.3, 0.4, 0.5, 0.7, 0.8, 0.9, 0.99, 1]
# IrradianceLoop = [0.3, 0.4, 0.5, 0.7, 0.8, 0.9, 0.99, 1]
# IrradianceLoop = [0.2, 0.5, 0.9, 1]
LoadMultLoop = [0.4, 0.6, 0.8, 1, 1.2]

for fault in FaultLabelSimulations:
    for irrads in IrradianceLoop:
        for Mults in LoadMultLoop:
            dss = py_dss_interface.DSS()
            opendss_path = "C:/Program Files/OpenDSS"
            dss = py_dss_interface.DSS(opendss_path)
            # dss_file = 'C:/OpenDSS_Basics/13Bus/IEEE13Nodeckt.dss'
            # dss_file = os.getcwd()
            # print(dss_file)
            #Compile, fault
            dss.text("compile {}".format(dss_file))
            irradiance = irrads 
            if irradiance != 0:
                
                #Objects needed to define PVSystem:
                dss.text("New XYCurve.Eff npts=4 xarray=[.1 .2 .4 1.0] yarray=[1 1 1 1]")
                dss.text("New XYCurve.FatorPvsT npts=4 xarray=[0 25 75 100] yarray=[1 1 1 1]") 
                dss.text("New Loadshape.Irrad npts=24 interval=1 mult=[0 0 0 0 0 0 .1 .2 .3 .5 .8 .9 1.0 1.0 .99 .9 .7 .4 .1 0 0 0 0 0]") 
                dss.text("New Tshape.Temp npts=24 interval=1 temp=[25 25 25 25 25 25 25 25 35 40 45 50 60 60 55 40 35 30 25 25 25 25 25 25]")

                 # irradiance = 1
                dss.text("New PVSystem.PV phases=3 bus1=890 Pmpp=" + str(270) +" kV=" + str(4.16) +" kVA=" + str(270) +" conn=wye effcurve=Eff P-TCurve=FatorPvsT %Pmpp=100 irradiance=" + str(irradiance) +" daily=Irrad Tdaily=Temp")
                dss.text("New PVSystem.PV2 phases=3 bus1=844 Pmpp=" + str(245) +" kV=" + str(24.9) +" kVA=" + str(245) +" conn=wye effcurve=Eff P-TCurve=FatorPvsT %Pmpp=100 irradiance=" + str(irradiance) +" daily=Irrad Tdaily=Temp")
            # dss.text("New PVSystem.PV3 phases=3 bus1=634 Pmpp=" + str(280) +" kV=" + str(0.48) +" kVA=" + str(280) +" conn=wye effcurve=Eff P-TCurve=FatorPvsT %Pmpp=100 irradiance=" + str(irradiance) +" daily=Irrad Tdaily=Temp")
            
            FaultEntered = fault[0]
            # dss.text("New Fault.FALL bus1=840 phases=3") #800
            # dss.text('New Fault.F1 phases=3 bus1=814 bus2=814.4.4.4 r=0.0001') #Neutral Example
            # print(FaultEntered)
            dss.text(FaultEntered) 
            dss.solution.load_mult = Mults
            
            # dss.text("Solve")
            dss.solution.solve()
            # if dss.solution.converged:
            #     print("Circuit Solved Successfully")
            # else:
            #     print("Circuit Not Solved")
            #     print(fault)
                    
            ConvergeSolution = dss.solution.converged
            
            filepath = dss.text("Export Voltages")
            # print(filepath)
            VoltageExport = pd.read_csv(filepath)
            VoltageExport["Fault_Command"] = FaultEntered
            VoltageExport["PV_Irradiance"] = irradiance
            # VoltageExport["Load_Multiple"] = LoadMultLoop
            VoltageExport["Fault_Type"] = fault[1]
            VoltageExport["CircuitSolved"] = ConvergeSolution
            FaultColumns = VoltageExport.columns
####################################################################################################333
            if VoltageExport.shape == (37,18):
                FaultSimulationData16 = np.concatenate((FaultSimulationData16,VoltageExport.to_numpy()), axis = 0)
            else:
                FaultSimulationData20 = np.concatenate((FaultSimulationData20,VoltageExport.to_numpy()), axis = 0)

faultdata16 = pd.DataFrame(FaultSimulationData16[1:], columns=FaultSimulationData16[0])
faultdata20 = pd.DataFrame(FaultSimulationData20[1:], columns=FaultSimulationData20[0])

faultdata20 = faultdata20[faultdata16.columns.to_list()]

AllFaultData = pd.concat([faultdata16,faultdata20])
AllFaultData = AllFaultData[AllFaultData['CircuitSolved']==True]

#Extract Classification Labels
AllFaultData['FaultLocation'] = AllFaultData['Fault_Command'].str.split().str[3].str.replace('bus1=','').str.split('.').str[0]

faultdata16.to_csv('C:/Users/ricky/Desktop/34Bus_project/Data/newest_faultdata16.csv', index = False)
faultdata20.to_csv('C:/Users/ricky/Desktop/34Bus_project/Data/newest_faultdata20.csv', index = False)
AllFaultData.to_csv('C:/Users/ricky/Desktop/34Bus_project/Data/newest_AllFaultData.csv', index = False)        
```
